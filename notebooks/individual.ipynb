```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Выделение объекта синего цвета
Загрузим изображение и конвертируем его к цветовой модели HSV (Hue, Saturation, Value — тон, насыщенность, значение) 

```python
img = cv2.imread ("catandfish.jpg" );
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

Теперь зададим границы синего цвета в HSV, а с помощью функции cv2.inRange() отфильтруем изображение по верхней и нижней границе синего цветаю. На выходе получается изображение из чёрных и белых пикслей (0 и 1). Далее функция cv2.bitwise_and() изменяет цвет неподходящих в изображении пикслеей на чёрный. А подходящие оставляет в исходном виде

```python
lower_blue = np.array([90, 50, 70])
upper_blue = np.array([130, 255, 255])
mask = cv2.inRange(hsv, lower_blue, upper_blue)
res = cv2.bitwise_and(img, img, mask = mask)

plt.subplot(221)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(222)
plt.imshow(cv2.cvtColor(mask,cv2.COLOR_BGR2RGB))
plt.axis("off")

plt.subplot(223)
plt.imshow(cv2.cvtColor(res,cv2.COLOR_BGR2RGB))
plt.axis("off")
```
